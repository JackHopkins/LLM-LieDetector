```python
import pickle

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

BINARY = True
```

```python
if BINARY:
    mode = "_binary"
else:
    mode = "" # logprobs

with open(f'results/modified_intention/averages_dfs_dict{mode}.pkl', 'rb') as f:
    modifiers_dfs_dict = pickle.load(f)
with open(f'results/refusals/averages_dfs_dict{mode}.pkl', 'rb') as f:
    refusal_dfs_dict = pickle.load(f)

for name in ["no_lie", "lie", "knowable", "subsets_union"]:
    
            
    # load modifiers_df
    modifiers_df = modifiers_dfs_dict[name]
    # load refusals_df
    refusal_df = refusal_dfs_dict[name]

    # who negates or enforces the initial lie instruction? Three cases: instruction, speaker (Answer: ), or speaker (Question: )
    # insert column called "Speaker" into modifiers_df which says "Question" whenever the index starts with "Question:", and "Instruction" otherwise
    modifiers_df['Speaker'] = np.where(modifiers_df["modifier"].str.startswith('Question:'), 'Question', 'Instruction')


    # edit refusal_df
    refusal_df['Speaker'] = "Answer"
    #rename column "Follow up lying rate" to Average Lying Rate
    # refusal_df.rename(columns={'Follow up lying rate': 'Average Lying Rate'}, inplace=True)

    #concat the dfs
    combined_df = pd.concat([modifiers_df, refusal_df], axis=0)
    combined_df.index.name = "modifier"

    display(combined_df)

    print("#########################")
    print(f"Elicitation question group: {name}")
    print("#########################")

    # Create a scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(combined_df['Average Prediction'], combined_df['Follow up lying rate'])
    plt.title("Scatter plot of average prediction vs average follow-up lying rate")
    plt.xlabel('Average Prediction')
    plt.ylabel('Follow up lying rate')

    # # Add labels to the points
    # for i, txt in enumerate(combined_df['modifier']):
    #     plt.annotate(txt, (combined_df['Average Prediction'].iat[i], combined_df['Follow up lying rate'].iat[i]))

    plt.show()

    # correlation between average prediction and follow-up lying rate
    print("Pearson's corr:\n", combined_df.corr(numeric_only=True))
    print("\n\n")
    print("Spearman's corr:\n", combined_df.corr(method="spearman", numeric_only=True))
```
